```python
from data_common.notebook import *
import json
import spacy
```

```python
items = json.loads(Path("data", "interim", "regmem.json").read_text())

df = pd.DataFrame(items).fillna("")
```

```python

class OrgExtractor:

    def __init__(self):
        self.nlp = spacy.load("en_core_web_sm")

    def get_orgs(self, text: str) -> list[str]:
        doc = self.nlp(text)
        return [x.text for x in doc.ents if x.label_ == "ORG"]

    def get_money(self, text: str) -> list[str]:
        doc = self.nlp(text)
        return [x.text for x in doc.ents if x.label_ == "MONEY" ]

oe = OrgExtractor()
df["extracted_orgs"] = df["free_text"].apply(oe.get_orgs).apply("; ".join)
df["extracted_money"] = df["free_text"].apply(oe.get_money).apply("; ".join)
df
```

```python
df.to_csv(Path("data", "data_packages", "regmem", "processed_regmem.csv"), index=False)
```
